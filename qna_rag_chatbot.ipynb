##Pdf loader

```python
from langchain_community.document_loaders import PyPDFLoader
```

```python
loader = PyPDFLoader("Machine-Learning-With-Python-For-Everyone-Pearson-2020.pdf")
```

```python
docs= loader.load()
len(docs)
```

```python
oneDoc = docs[48]
print(oneDoc.page_content)
```

##Splitter

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
splitter = RecursiveCharacterTextSplitter(chunk_size = 1000, chunk_overlap = 200)
```

```python
pdf_chunks = splitter.split_documents(docs)
len(pdf_chunks)
```

```python
print(pdf_chunks[48].page_content)
```

##Vector Embedding 


```python
from langchain_huggingface import HuggingFaceEmbeddings
```

```python
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-mpnet-base-v2")
```

##Vector stores

```python
from langchain_community.vectorstores import InMemoryVectorStore
```

```python
vector_db = InMemoryVectorStore.from_documents(
        documents=docs,
        embedding=embeddings
    ) #inmemory vector store    
```

##Stores in vector dbs

```python
from langchain_chroma import Chroma
```

```python
vector_store = Chroma.from_documents(
    documents=docs,
    embedding=embeddings,
    persist_directory="./vector_db"
)
```

```python
query = "what is machine learning?"
```

```python
result = vector_store.similarity_search(query, k=3)
```

```python
result[0].page_content
```
